```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\MachineLearning\dataset\FuelConsumption.csv");
df.head()
```

```python
df.columns
```

```python
x=df['FUELCONSUMPTION_COMB_MPG'].values
y=df['CO2EMISSIONS'].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
xtrain=xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
xtest=xtest.reshape(-1,1)
xtest.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
pipe=make_pipeline(StandardScaler(),LinearRegression())
pipe.fit(xtrain,ytrain)
y_pred=pipe.predict(xtest)
pipe
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.datasets import load_iris
x=load_iris(return_X_y=False)
param={
    'kernel':['rbf','poly'],'C':[0.003,0.05,0.3,0.001],'gamma':[0.1,0.2]
}
search=GridSearchCV(SVC(),param,cv=2)
search.fit(x.data,x.target)
s=search.predict(x.data)
search.cv_results_
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.datasets import load_iris
x=load_iris(return_X_y=False);
pipe=Pipeline(steps=[('std',StandardScaler()),('pca',PCA()),('svc',SVC())])
param={
    'svc__kernel':['rbf','poly','linear'],'svc__C':[0.1,0.6,0.4],"pca__n_components":[2,3]
}
search=GridSearchCV(pipe,param,cv=5)
search.fit(x.data,x.target)
search.predict(x.data)
search.cv_results_
```
